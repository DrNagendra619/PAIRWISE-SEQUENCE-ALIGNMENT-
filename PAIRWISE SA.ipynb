```python
from Bio import Align
Aligner= Align.PairwiseAligner(match_score = 1.0)
```

```python
SeqA = "GAATCCAG"
SeqB = "AACAG"

score = Aligner.score(SeqA, SeqB)
```

```python
print(f"The Alignment score is {score}")
```

```python
Alignments = Aligner.align(SeqA, SeqB)

for Alignment in Alignments:
    print(Alignment)
```

```python
Align.mode = "local"
Seq1 = "ATTGCGATTAAA"
Seq2 = "GCG"
```

```python
Local_score = Aligner.score(Seq1, Seq2)
print(f"The local alignment score is {Local_score}")
```

```python
Align_Local = Aligner.align(Seq1, Seq2)
for LA in Align_Local:
    print(LA)
```
